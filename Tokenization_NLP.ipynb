```python
msg= '''
hi, how are you ? 
yeah I am talking to you.
yes you bloody Idiot,
seriously still no reply.
ok get the hell out of here.
'''
```

```python
# import tokenizer
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
```

```python
# sentence tokenize
sent=sent_tokenize(msg)
```

```python
sent
```

```python
# no of sent
len(sent)
```

```python
# now tokenizing by words
word = word_tokenize(msg)
```

```python
len(word)
```

```python
# stopwords remove
from nltk.corpus import stopwords
```

```python
# remove stop words from actual words
new_word=[i for i in word if i.lower() not in stopwords.words('english')]
```

```python
len(new_word)
```

```python
# freq
import nltk
import matplotlib.pyplot as plt
```

```python
nlp=nltk.FreqDist(word)
nlp.plot()  # plotting actual Words
```

```python
nlp1=nltk.FreqDist(new_word)
nlp1.plot()  # plotting new words
```
