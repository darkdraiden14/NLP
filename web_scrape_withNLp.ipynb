```python
# reading from file
f = open('myweb.txt','r')
mydata=f.read()
f.close()
```

```python
type(mydata)
```

```python
# split
newdata1 = [i for i in mydata.split()]
```

```python
# NLTK graph
import matplotlib.pyplot as plt
import nltk
```

```python
[i for i in dir(nltk) if 'Freq' in i]
```

```python
nlpdata=nltk.FreqDist(newdata1)
```

```python
nlpdata.plot(20)
```

```python
# we can remove stopwords from newdata1 then plot the graph
from nltk.corpus import stopwords
```

```python
# removing stopwords
removedata=[i for i in newdata1 if i not in stopwords.words('english')]
```

```python
# apply NLTK FREQDIST
nlpremove = nltk.FreqDist(removedata)
```

```python
nlpremove.plot(20)
```
